## Boston Crime: When Danger Peaks & Where It Lurks

This notebook fetches crime incident data from the **Analyze Boston SQL API** and processes it into two aggregated datasets for our Plotly.js visualizations:

1. **Heatmap data** — Crime counts by Day of Week × Hour
2. **Grouped bar data** — Top offense categories by District

```python
import urllib
import requests
import pandas as pd
import json
```

Fetch Data from Analyze Boston SQL API

Using the same approach from class — the `datastore_search_sql` endpoint with the crime incident resource ID.

```python
# API endpoint (same as class example)
bpd_logs = "https://data.boston.gov/api/3/action/datastore_search_sql"

# Resource ID for crime incident reports
RESOURCE_ID = "b973d8cb-eeb2-4e7e-99da-c92938efc9c0"
```

```python
# Query 1: Get all records with the fields we need for our story
sql = f'''
SELECT "DISTRICT",
       "OCCURRED_ON_DATE",
       "OFFENSE_CODE_GROUP",
       "OFFENSE_DESCRIPTION",
       "DAY_OF_WEEK",
       "HOUR",
       "MONTH",
       "YEAR",
       "SHOOTING",
       "STREET"
FROM "{RESOURCE_ID}"
LIMIT 30000
'''

url = bpd_logs + "?sql=" + urllib.parse.quote(sql)
rows = requests.get(url).json()
records = rows['result']['records']

print(f"Fetched {len(records)} records")
print(f"Sample record keys: {list(records[0].keys()) if records else 'No records'}")
```

```python
# Convert to DataFrame
df = pd.DataFrame(records)
print(f"DataFrame shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
df.head()
```
